```python
# import numpy as np
import matplotlib.pyplot as plt

import pyrow.physical_model as physical_model
import pyrow.polar_diagram as polar_diagram
```

```python
# estimates Rannoch R45
boat_properties = physical_model.BoatProperties(
    area_anchor=6,
    area_water_front=0.329,
    area_water_side=1.87 + 0.196 + 0.184,
    area_air_front=1.89,
    area_air_side=5.81,
    drag_coefficient_air=0.2,
    drag_coefficient_water=0.0075,
    speed_perfect=4,
    min_anchor_speed=None,
    wind_correction={"correction_par": 2, "window_par": 6, "window_perp": 4},
)

# Polar diagram
v_air_perp_mat, v_air_par_mat, vperp_mat_boat, vpar_mat_boat = (
    polar_diagram.calculate_mat_polar_diagram(boat_properties)
)
fig, ax = polar_diagram.plot_polar_diagram(
    boat_properties, v_air_perp_mat, v_air_par_mat, vpar_mat_boat
)
plt.show()


# Polar diagram difference row-rest
v_air_perp_mat, v_air_par_mat, vperp_mat_boat_diff, vpar_mat_boat_diff = (
    polar_diagram.calculate_boat_velocity_difference_rest_row(boat_properties)
)
fig, ax = polar_diagram.plot_polar_diagram(
    boat_properties,
    v_air_perp_mat,
    v_air_par_mat,
    vpar_mat_boat_diff,
    contour_step=0.1,
    vmin=0,
    cmap="viridis",
    title="Boat Velocity Polar Diagram Row-Rest",
    draw_row_zone=True,
)
plt.show()
```
